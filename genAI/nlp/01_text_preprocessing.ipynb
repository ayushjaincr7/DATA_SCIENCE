#3. url remover

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub("", text)
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char, "")
  return text
```

```python
def remove_punc1(text):
  return text.translate(str.maketrans("","", exclude))
```

# 5. remove slang

```python
text="FYI this is not true"
text2="LAMO the class was so funny"
text3="i want report ASAP"
```

```python
chat_words={
   "AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LAMO":"Laugh My A.. Off",
"FYI":"For your information"
}
```

```python
chat_words["AFAIK"]
```

```python
def chat_conversion(text):
  new_text = []
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
chat_conversion(text)
```

# 7. stop word removing

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append("")
    else:
      new_text.append(word)

  return " ".join(new_text).replace("   ","")
```

```python
text="i am sunny and we are learning nlp and here is why nlp are important"
```

```python
remove_stopwords(text)
```

```python
import re
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', text)
    return clean_text
```

```python
print(remove_emojis_manually(text4))
```

```python
print(remove_emojis_manually(original_text))
```

```python
import spacy

# Load the English NLP model from spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def tokenize_text(text):
    # Process the text with the spacy model
    doc = nlp(text)

    # Extract tokens from the processed document
    tokens = [token.text for token in doc]

    return tokens
```

```python
tokens=tokenize_text(my_corpus)
```

```python
for word in tokens:
    print(word)
```

```python


# Example usage:
text = "Spacy is a powerful natural language processing library."
tokens = tokenize_text(text)

print(f"Original text: {text}")
print(f"Tokens: {tokens}")
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```

```python
from nltk.stem import WordNetLemmatizer

def lammatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=[lemmetizer.lemmatize(word) for word in words]

    return lemetized_word
```
